#### 2.2.2 Price outlier inspect

```python
def highlight_extreme_outliers(value):
    '''
    highlight the extreme in a Series yellow.
    '''
    is_extreme = value > 8000
    return ['background-color: red' if v else '' for v in is_extreme]
```

```python
# keep only the ones that are out of +3 to -3 standard deviations in the column 'price'.
price_outliers = df[~(np.abs(df['price']-df['price'].mean())<(3*df['price'].std()))].sort_values('price', ascending=False) 
price_outliers.style.apply(highlight_extreme_outliers, subset=['price'])
```

#### 2.2.3 Important features to price

```python
def plot_feature_importances(dataset_name, features, importances, indices):
    """Plot the 15 most important features."""
    a = 0.8
    f_sub = []
    max_features = 20

    for f in range(min(len(features), max_features)): 
            f_sub.append(f)

    # Create a figure of given size
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111)
    # Set title
    ttl = dataset_name

    df = pd.DataFrame(importances[indices[f_sub]][::-1])
    df.plot(kind='barh', ax=ax, alpha=a, legend=False, edgecolor='w', 
            title=ttl, color = [plt.cm.viridis(np.arange(len(df))*10)])

    # Remove grid lines and plot frame
    ax.grid(False)
    ax.set_frame_on(False)

    # Customize title
    ax.set_title(ax.get_title(), fontsize=14, alpha=a, ha='left', x=0, y=1.0)
    plt.subplots_adjust(top=0.9)

    # Customize x tick lables
    ax.xaxis.set_major_formatter(FormatStrFormatter('%.3f'))
    ax.locator_params(axis='x', tight=True, nbins=5)

    # Customize y tick labels
    yticks = np.array(features)[indices[f_sub]][::-1]
    ax.set_yticklabels(yticks, fontsize=12, alpha=a)
    ax.yaxis.set_tick_params(pad=2) 
    ax.yaxis.set_ticks_position('none') 
    ax.set_ylim(ax.get_ylim()[0]-0.5, ax.get_ylim()[1]+0.5) 

    # Set x axis text
    xlab = 'Feature importance'
    ax.set_xlabel(xlab, fontsize=12, alpha=a)
    ax.xaxis.set_label_coords(0.5, -0.1)

    # Set y axis text
    ylab = 'Feature'
    ax.set_ylabel(ylab, fontsize=12, alpha=a)
    plt.show()
```

```python
y = df_numeric['price']
X = df_numeric[df_numeric.columns.difference(['price'])]
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.8, random_state = 42)
```

```python
rf = RandomForestRegressor(n_estimators = 100,
                           max_depth = 20,
                           n_jobs = -1,
                           oob_score = True,
                           bootstrap = True,
                           random_state = 42)
rf.fit(X_train, y_train)

# print('R^2 Training Score: {:.2f} \nOOB Score: {:.2f} \nR^2 Validation Score: {:.2f}'.format(rf.score(X_train, y_train), 
#                                                                                              rf.oob_score_,
#                                                                                              rf.score(X_valid, y_valid)))
```

```python
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]
plot_feature_importances("Feature importantce to price", X.columns, importances, indices)
```

### 3.1 Municipality

#### 3.1.1 Violations

In principle, holiday rental is only allowed if you are the owner of the house or houseboat.As a tenant, you can almost never rent your house or houseboat to others. <br>
You may only rent the house or houseboat occasionally 60 (2016) <br>
You can’t rent out your house or houseboat to more than four people at the same time.

#### 60 days limit

```python
# exceed 60 days rental limit
df_apt = df[df['room_type']=='Entire home/apt']
df_apt_old = df_apt[df_apt['number_of_reviews']>60]
df_exceed60 = df_apt_old[df_apt_old['reviews_per_month']>5]
df_exceed60 = df_exceed60[['name', 'room_type', 'price', 'number_of_reviews', 'reviews_per_month']]
df_exceed60
```

```python
df_exceed60['extra_days_per_year'] = df_exceed60['reviews_per_month'] * 12 - 60
df_exceed60['estimated_extra_income'] = df_exceed60['extra_days_per_year'] * df_exceed60['price'] * 0.8
df_exceed60
```
